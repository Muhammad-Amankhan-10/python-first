```python
import random

game = ["rock", "paper", "scissors"]
flag = True

while flag:
    Ai_player = game[random.randint(0, 2)]
    player_1 = input(f"Enter your choice {game}: ").lower()
    
    print("Player 1 choice:", player_1)
    print("Ai_player choice:", Ai_player)
    
    if player_1 == Ai_player:
        print("It's a tie! Let's play again.")
    elif (player_1 == "rock" and Ai_player == "scissors") or \
         (player_1 == "paper" and Ai_player == "rock") or \
         (player_1 == "scissors" and Ai_player == "paper"):
        print("Player 1 wins!")
    else:
        print("Ai_player wins!")

    play_again = input("You wanna play again (y/n)? ").lower()
    if play_again == "n":
        flag = False
        print("Thanks for playing!")
```

```python
import random

print("ðŸŽ¯ Welcome to the Number Guessing Game!")

# Ask user for difficulty level
difficulty = input("Choose difficulty (easy/hard): ").lower()

if difficulty == "easy":
    number = random.randint(1, 50)
    print("You chose EASY mode! Guess a number between 1 and 50.")
elif difficulty == "hard":
    number = random.randint(50, 100)
    print("You chose HARD mode! Guess a number between 50 and 100.")
else:
    print("Invalid choice! Defaulting to EASY mode.")
    number = random.randint(1, 50)

# Initialize attempts
attempts = 0
max_attempts = 10

# Start guessing loop
while attempts < max_attempts:
    guess = int(input("Enter your guess: "))
    attempts += 1

    if guess > number:
        print("Too high! ðŸ“ˆ")
    elif guess < number:
        print("Too low! ðŸ“‰")
    else:
        print(f"ðŸŽ‰ Correct! You guessed it in {attempts} attempts.")
        break  # exit loop if correct

    # Remaining attempts info
    print(f"Attempts left: {max_attempts - attempts}")

else:
    print(f"âŒ Game over! The correct number was {number}.")
```
